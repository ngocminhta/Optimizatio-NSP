```python
from ortools.sat.python import cp_model

with open('1.txt') as file:
  n, d, a, b = [int(q) for q in file.readline().split()]
  f = [[0 for j in range(d)] for i in range(n)]
  for i in range(n):
    for j in [int(h) for h in file.readline().split()]:
      if j != -1:
            f[i][j-1] = 1

model = cp_model.CpModel()
x = [[[ None for k in range(4)] for j in range(d)] for i in range(n)]
for i in range(n):
    for j in range(d):
        for k in range(4):
            x[i][j][k] = model.NewIntVar(0, 1, f'x_{i}_{j}_{k}')
t = model.NewIntVar(0, d, 't')

# moi ngay 1 ca
for i in range(n):
    for j in range(d):
        p = 0
        for k in range(4):
            p+= x[i][j][k]
        model.Add(p <= 1)

#toi thieu va toi da
for j in range(d):
    for k in range(4):
        p = 0
        for i in range(n):
            p += x[i][j][k]
        model.Add(p >= a)
        model.Add(p <= b)

for i in range(n):
    p = 0
    for j in range(d):
        p += x[i][j][3]
    model.Add(t - p >= 0)

# k phai ngay nghi thi phai di llam
for i in range(n):
    for j in range(1,d):
        p = 0
        for k in range(4):
            p += x[i][j][k]
        model.Add(p + x[i][j-1][3] >= 1-f[i][j])

j = 0
for i in range(n):
    p = 0
    for k in range(4):
        p += x[i][j][k]
    model.Add(p >= 1 - f[i][j])

model.Minimize(t)
solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL:
    print(f'Minimum of objective function: {solver.ObjectiveValue()}\n')
    for i in range(n):
        for j in range(d):
            for k in range(4):
                x[i][j][k] = solver.Value(x[i][j][k])
    for i in range(n):
        print("Người thứ",i+1)
        for j in range(d):
            print('Ngày thứ',j+1,':',*x[i][j])
```
